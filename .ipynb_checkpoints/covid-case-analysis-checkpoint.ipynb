```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
plt.style.use('fivethirtyeight')
```

```python
df = pd.read_csv('data/India-District-Zones.csv')
```

```python
df
```

# Percentage of Green, Red and Orange zone in India

```python
color = ["#008000","#FFA500","#FF0000"]
df['Zone'].value_counts(normalize=True).plot(kind='bar',color=color)
```

```python
state_grp = df.groupby(['State'])
```

```python
states = []
for state in state_grp:
    states.append(state[0])
print(states)
```

```python
sta = []
for state in states:
    add=dict(state_grp.get_group(state)['Zone'].value_counts())
    sorted_dict = dict(sorted(add.items()))
    sta.append(sorted_dict)
    country = dict(zip(states,sta))
    
print(len(country))  
```

```python
orange = []
for state in country:
    
    if 'Orange Zone' in country[state]:
        orange.append(country[state]['Orange Zone'])
    else:
        orange.append(0)

print(orange)
```

```python
red = []
for state in country:
    
    if 'Red Zone' in country[state]:
        red.append(country[state]['Red Zone'])
    else:
        red.append(0)

print(red)
```

```python
green = []
for state in country:
    
    if 'Green Zone' in country[state]:
        green.append(country[state]['Green Zone'])
    else:
        green.append(0)

print(green)
```

# Number of Green zone in each states

```python
plt.barh(states,green,color="#008000",label="Green Zone")
#plt.bar(x_indexes,orange,width=width,color="#FFA500",label="Orange Zone")
#plt.bar(x_indexes+width,red,width=width,color="#FF0000",label="Red Zone")
plt.legend()
plt.tight_layout()
plt.ylabel("States")
plt.rcParams['figure.figsize'] = [50/2.54, 40/2.54]
plt.xlabel("Number of Zones")
plt.show()
```

## Number of Ornage zone in each states

```python
plt.barh(states,orange,color="#FFA500",label="Orange Zone")
plt.legend()
plt.tight_layout()
plt.ylabel("States")
plt.rcParams['figure.figsize'] = [50/2.54, 40/2.54]
plt.xlabel("Number of Zones")
plt.show()
```
